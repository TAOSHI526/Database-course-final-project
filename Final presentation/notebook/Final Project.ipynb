```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

%matplotlib inline
```

```python
response = requests.get('http://www.bbc.com/culture/story/20160819-the-21st-centurys-100-greatest-films-who-voted')
doc = BeautifulSoup(response.text)
```

```python
all_p = doc.find_all('p')
```

```python
all_p[4]
```

```python
critics = doc.find_all('strong')
critics
```

```python
# row['critic_info'] = p.strong.string
movies=[]
authors=[]
for p in all_p: 
    if p.strong is not None:
        critic = p.strong.string
        info = re.split("\d\.\s", p.next_sibling.text)
        for movie in info[1:]:
            movies.append(movie) 
            authors.append(critic)
```

```python
movie_info = pd.DataFrame({'critic_info':authors,
              'movie':movies                
})
```

```python
movie_info.head()
```

```python
movie_info
```

```python
movie_info['movie_name']=movie_info.movie.str.extract(r'^(.*)\(')
```

```python
movie_info['movie_director']=movie_info.movie.str.extract(r'\((.*)\,\s\d')
```

```python
movie_info['movie_year']=movie_info.movie.str.extract(r'\,\s(\d+)\)')
```

```python

rows=[]
for p in all_p:
    row={}
    if p.strong is not None:
        row['critic_info'] = p.strong.string
        row['movie_info'] = p.next_sibling.text.strip()
        rows.append(row)
df = pd.DataFrame(rows)
df
```

```python
df = df.iloc[0:-1,:]
df
```

```python
df['critic'] = df.critic_info.str.extract(r'(^[A-Z].+\s[A-Z][a-z]+.*) – ')
```

```python
df['org'] = df.critic_info.str.extract(r'–\s([A-Z].*)\s\(')
```

```python
df['country'] = df.critic_info.str.extract(r'\((.*)\)$')
```

```python
director_ids=[]
for url in search_url:
    response2 = requests.get(url)
    doc = BeautifulSoup(response2.text)
    try: 
        director_ids.append(doc.find('tr', attrs={'class':"findResult odd"}).find('a').attrs['href'])
    except:
        director_ids.append(None)
```

```python
director_ids
```

```python
directors['d1_url']=director_ids
```

```python
import time
```

```python
d_births = []
birth_places = []
for director in directors['d1_url']:
    url = 'https://www.imdb.com' + director
    try: 
        response3 = requests.get(url)
        doc = BeautifulSoup(response3.text)
        d_births.append(doc.find('time').find_all('a')[1].string)
        birth_places.append(doc.find('div', attrs={'id':"name-born-info"}).find_all('a')[-1].string)
    except:
        d_births.append(None)
        birth_places.append(None)
    time.sleep(0.5)
```
